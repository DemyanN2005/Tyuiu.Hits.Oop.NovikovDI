Описание проекта
Проект представляет собой систему управления проектами в рамках организации. Основная цель проекта — обеспечить структурированное управление различными типами проектов, такими как продуктовые, исследовательские и инфраструктурные проекты. Система позволяет отслеживать статус проектов, назначать командных членов и получать детальную информацию о проектах.

Основные задачи, которые решает проект:
Управление проектами: Система позволяет создавать и управлять различными типами проектов, отслеживать их статус и детали.
Назначение командных членов: Возможность назначать членов команды на проекты с учетом их специфических навыков и ролей.
Отслеживание статуса: Обновление и отслеживание статуса проектов, включая ожидаемую дату завершения и другие важные атрибуты.
Получение деталей проекта: Возможность получения подробной информации о проекте, включая его описание, статус, командных членов и другие атрибуты.
Примененные методы и подходы:
Наследование:

Базовый класс Project: Содержит общие атрибуты и методы, такие как ProjectId, ProjectName, Status, Description, UpdateStatus, GetProjectDetails, AssignTeamMember.
Производные классы:
ProductProject: Добавляет атрибуты ExpectedCompletionDate и ProductType, переопределяет метод UpdateStatus.
ResearchProject: Добавляет атрибуты FundingAmount и ResearchField, переопределяет метод AssignTeamMember.
InfrastructureProject: Добавляет атрибуты ExecutionPeriod и Location, переопределяет метод GetProjectDetails.
Полиморфизм:

Переопределение методов в производных классах для демонстрации полиморфизма. Например, метод UpdateStatus в ProductProject включает ожидаемую дату завершения.
Использование перегрузки методов для различных сценариев. Например, метод AssignTeamMember в базовом классе имеет перегрузку для назначения члена команды с указанием роли.
Интерфейсы:

Реализация интерфейса IProject для обеспечения явной реализации методов и управления зависимостями.
Коллекции:

Использование списка List<string> для хранения командных членов проекта.
Делегаты и события:

Реализация события TeamMemberAssigned для уведомления о назначении нового члена команды.
Generic классы:

Использование обобщенного класса GenericProject<T> для хранения и отображения данных проекта.

Заключение
Проект демонстрирует использование основных принципов объектно-ориентированного программирования, таких как наследование, полиморфизм, интерфейсы, коллекции, делегаты и события. Он предоставляет гибкую и расширяемую систему для управления проектами в организации, обеспечивая возможность легко добавлять новые типы проектов и функциональности.
